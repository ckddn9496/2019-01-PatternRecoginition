```python
import matplotlib.pyplot as plt
import numpy as np
```

$$f(x) = x^2 - 2x + 1 \; (-10 \leq x \leq 10)$$
$$f'(x) = 2x - 3 \; ( x = 2 )$$
$$x : [-10,10]$$
$$ y : [-40,40]$$

```python
# Define a differentiable funcion that maps from real number to read number.
plt.axis([-10,10,-40,40])

# Define a domain of the function.
x = np.arange(-10.0, 10.0, 0.01)

# Plot the function.
plt.plot(x, x**2-2*x+1, 'r', label='y=x^2-2x+1')
# Select a point within the domain.
px = np.array([2])
py = np.array([1])

# Mark the selected point on the function.
plt.scatter(px, py)

# Define the first-order Taylor approximation at the selected point.
plt.plot(x, 2*x-3, 'b--', label='y=2x-3')

plt.title('Assignment 2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Plot the Taylor approximation with the same domain of the original function.
plt.show()
```
