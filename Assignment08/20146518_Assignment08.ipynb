```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

<h2>choose a polynomial curve(y1) and generate points along the curve with random noise(y2)</h2>

```python
num = 1001
std = 2
n = np.random.rand(num)
nn = n - np.mean(n)
x = np.linspace(-10,10,num)
y1 = np.sin(x) + np.sinc(x)
y2 = y1 + nn * std
```

```python
print(x)
print(y1)
print(y2)
```

<h2>plot the generated noisy points along with its original polynomial without noise</h2>

```python
plt.plot(x, y2, 'k.', x, y1, 'r.')
plt.show()
```

<h2>plot the approximating polynomial curve obtained by solving a least square problem using</h2>
$$x=(A^TA)^{-1}A^Tb$$

```python
A = np.zeros((num,2))
for i in range(num):
    A[i,0]=1
    A[i,1]=x[i]

theta = ((np.linalg.inv(A.T.dot(A))).dot(A.T)).dot(y1)
plt.plot(x, theta[0]+theta[1]*x, 'k.', x, y1, 'r.')
plt.show()
```
