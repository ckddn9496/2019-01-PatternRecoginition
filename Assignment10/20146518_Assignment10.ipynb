<h2>Build a binary classifier for each digit against all the other digits at MNIST dataset.</h2>

```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>load train and test datas from .csv files</h2>

```python
file_data_train = "mnist_train.csv"
file_data_test  = "mnist_test.csv"

h_data_train    = open(file_data_train, "r")
h_data_test     = open(file_data_test, "r")

data_train      = h_data_train.readlines()
data_test       = h_data_test.readlines()

h_data_train.close()
h_data_test.close()

size_row    = 28    # height of the image
size_col    = 28    # width of the image

num_train   = len(data_train)   # number of training images
num_test    = len(data_test)    # number of testing images

#
# normalize the values of the input data to be [0, 1]
#
def normalize(data):

    data_normalized = (data - min(data)) / (max(data) - min(data))

    return(data_normalized)

#
# example of distance function between two vectors x and y
#
def distance(x, y):

    d = (x - y) ** 2
    s = np.sum(d)
    # r = np.sqrt(s)

    return(s)

#
# make a matrix each column of which represents an images in a vector form 
#
list_image_train    = np.empty((size_row * size_col, num_train), dtype=float)
list_label_train    = np.empty(num_train, dtype=int)

list_image_test     = np.empty((size_row * size_col, num_test), dtype=float)
list_label_test     = np.empty(num_test, dtype=int)

count = 0
# read train data
for line in data_train:

    line_data   = line.split(',')
    label       = line_data[0]
    im_vector   = np.asfarray(line_data[1:]) #float형으로 변형
    im_vector   = normalize(im_vector)

    list_label_train[count]     = label
    list_image_train[:, count]  = im_vector    
   
    count += 1

count = 0
# read test data
for line in data_test:
    line_data   = line.split(',')
    label       = line_data[0]
    im_vector   = np.asfarray(line_data[1:])
    im_vector   = normalize(im_vector)

    list_label_test[count]      = label
    list_image_test[:, count]   = im_vector    

    count += 1
```

```python
def multi_label_classifier_with_randomfeatures(p):
    #-------------------------------------------------------------------------------
    # finding out model parameters(theta) for each digit
    list_label_train_ = np.empty(num_train, dtype=int)
    # read train data and set 1 or -1 wher label is not correct
   
    R=np.empty((p,size_row * size_col),dtype=float)
    count=0
    for i in range(p): # make random image and set on matrix R
        r=np.random.normal(0,1,size_row * size_col)
        R[count,:]  = r    
        count += 1
        
    A_r=(R.dot(list_image_train)).T
    
    # compute theta_d_r
    for i in range(num_train):
        if (list_label_train[i]==0):
            list_label_train_[i]=1
        else:
            list_label_train_[i]=-1
    theta0_r=((np.linalg.pinv((A_r.T).dot(A_r))).dot(A_r.T)).dot(list_label_train_)
    for i in range(num_train):
        if (list_label_train[i]==1):
            list_label_train_[i]=1
        else:
            list_label_train_[i]=-1
    theta1_r=((np.linalg.pinv((A_r.T).dot(A_r))).dot(A_r.T)).dot(list_label_train_)
    
    for i in range(num_train):
        if (list_label_train[i]==2):
            list_label_train_[i]=1
        else:
            list_label_train_[i]=-1
    theta2_r=((np.linalg.pinv((A_r.T).dot(A_r))).dot(A_r.T)).dot(list_label_train_)
    
    for i in range(num_train):
        if (list_label_train[i]==3):
            list_label_train_[i]=1
        else:
            list_label_train_[i]=-1
    theta3_r=((np.linalg.pinv((A_r.T).dot(A_r))).dot(A_r.T)).dot(list_label_train_)
    
    for i in range(num_train):
        if (list_label_train[i]==4):
            list_label_train_[i]=1
        else:
            list_label_train_[i]=-1
    theta4_r=((np.linalg.pinv((A_r.T).dot(A_r))).dot(A_r.T)).dot(list_label_train_)
    
    for i in range(num_train):
        if (list_label_train[i]==5):
            list_label_train_[i]=1
        else:
            list_label_train_[i]=-1
    theta5_r=((np.linalg.pinv((A_r.T).dot(A_r))).dot(A_r.T)).dot(list_label_train_)
    
    for i in range(num_train):
        if (list_label_train[i]==6):
            list_label_train_[i]=1
        else:
            list_label_train_[i]=-1
    theta6_r=((np.linalg.pinv((A_r.T).dot(A_r))).dot(A_r.T)).dot(list_label_train_)
    
    for i in range(num_train):
        if (list_label_train[i]==7):
            list_label_train_[i]=1
        else:
            list_label_train_[i]=-1
    theta7_r=((np.linalg.pinv((A_r.T).dot(A_r))).dot(A_r.T)).dot(list_label_train_)
    
    for i in range(num_train):
        if (list_label_train[i]==8):
            list_label_train_[i]=1
        else:
            list_label_train_[i]=-1
    theta8_r=((np.linalg.pinv((A_r.T).dot(A_r))).dot(A_r.T)).dot(list_label_train_)
    for i in range(num_train):
        if (list_label_train[i]==9):
            list_label_train_[i]=1
        else:
            list_label_train_[i]=-1
    theta9_r=((np.linalg.pinv((A_r.T).dot(A_r))).dot(A_r.T)).dot(list_label_train_)
    
    #------------------------------------------------------------------
    
    #---this section is classifying the digits in train set -----------
    y_train=np.empty((10,num_train),dtype=float) # make container
    ftrain=(R.dot(list_image_train)).T
    y_train[0,:]=ftrain.dot(theta0_r)
    y_train[1,:]=ftrain.dot(theta1_r)
    y_train[2,:]=ftrain.dot(theta2_r)
    y_train[3,:]=ftrain.dot(theta3_r)
    y_train[4,:]=ftrain.dot(theta4_r)
    y_train[5,:]=ftrain.dot(theta5_r)
    y_train[6,:]=ftrain.dot(theta6_r)
    y_train[7,:]=ftrain.dot(theta7_r)
    y_train[8,:]=ftrain.dot(theta8_r)
    y_train[9,:]=ftrain.dot(theta9_r)
    
    label_about_train=np.empty(num_train,dtype=int)
    # find max prediction value's label and save into label_about_train
    for i in range(num_train):
        max_=y_train[0,i]
        label_about_train[i]=0
        for j in range(1,10):
            if(y_train[j,i]>max_):
                label_about_train[i]=j
                max_=y_train[j,i]
                
    # generate confusion train container
    confusion_train=np.zeros((10,10))
    
    for i in range(num_train):
        # confusion_train[f][f^] => f: label, f^: predict
        confusion_train[list_label_train[i],label_about_train[i]]+=1
    print('\n\n\n---------------------------------------',)
    print('when p=',p)
    print('\nabout train set confusion matrix')
    for i in range(10):
        print()
        for j in range(10):
            print('   ',confusion_train[i][j],end='')
    c=0
    for i in range(10):
        c+=confusion_train[i][i]
    accuracy = c/num_train
    print("\ntrue positive rate about train set: ", accuracy)
    print("\nerror rate about train set: ", 1-accuracy)
    
    #------------------------------------------------------------------
    
    #--------------------- same in task set ---------------------------
    y_test=np.empty((10,num_test),dtype=float)
    ftest=(R.dot(list_image_test)).T
    y_test[0,:]=ftest.dot(theta0_r)
    y_test[1,:]=ftest.dot(theta1_r)
    y_test[2,:]=ftest.dot(theta2_r)
    y_test[3,:]=ftest.dot(theta3_r)
    y_test[4,:]=ftest.dot(theta4_r)
    y_test[5,:]=ftest.dot(theta5_r)
    y_test[6,:]=ftest.dot(theta6_r)
    y_test[7,:]=ftest.dot(theta7_r)
    y_test[8,:]=ftest.dot(theta8_r)
    y_test[9,:]=ftest.dot(theta9_r)
    label_about_test=np.empty(num_test,dtype=int)
    for i in range(num_test):
        max_=y_test[0,i]
        label_about_test[i]=0
        for j in range(1,10):
            if(y_test[j,i]>max_):
                label_about_test[i]=j
                max_=y_test[j,i]
    confusion_test=np.zeros((10,10))
    for i in range(num_test):
        confusion_test[list_label_test[i],label_about_test[i]]+=1
    print('\n\nabout test set confusion matrix')
    for i in range(10):
        print()
        for j in range(10):
            print('   ',confusion_test[i][j],end='')
    c=0
    for i in range(10):
        c+=confusion_test[i][i]
    accuracy = c/num_test
    print("\ntrue positive rate test set: ",c/num_test)
    print("\nerror rate about train set: ", 1-accuracy)
    #-----------------------------------------------------------------
    return
```
